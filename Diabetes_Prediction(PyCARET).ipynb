```python
from pycaret.datasets import get_data
data = get_data('diabetes')
```

```python
from pycaret.classification import *
s = setup(data, target = 'Class variable')
```

```python
best_model=compare_models()
```

```python
print(best_model)
```

```python
logistic_regression_model=create_model('lr', fold = 20)
```

```python
plot_model(logistic_regression_model,'learning')
```

```python
tuned_logistic_regression=tune_model(logistic_regression_model,optimize='Accuracy')
```

```python
plot_model(logistic_regression_model,'learning')
```

```python
plot_model(logistic_regression_model,'auc')
```

```python
save_model(tuned_logistic_regression,'diabetes_pred')
```
